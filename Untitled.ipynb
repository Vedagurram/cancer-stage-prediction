```python
import numpy as np
from sklearn import preprocessing, neighbors, svm
from sklearn.model_selection import cross_validate
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("breast-cancer-wisconsin.data")
```

```python
df
```

```python
# replacing the missing data in dataframe with -99999
df.replace("?", -99999, inplace=True)
```

```python
# dropping the id column as it useless..
df.drop(['Id'], 1, inplace=True)
```

```python
df
```

```python
# defining X(features) and y(lables)
X = np.array(df.drop(['class'],1))  # features are everything except the "class" column.
y = np.array(df['class'])   # lable is only the class column.
```

```python
# doing cross_validation

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
# the function above will take our features and labels and shuffle them up and splits the data according to ratio, and
# returns the values
```

```python
# defining the model

model = svm.SVC()
```

```python
model.fit(X_train, y_train)
```

```python
accuracy = model.score(X_test, y_test)
```

```python
print(accuracy)
```

```python
# now we are predicting some of the examples:-
example_measures = np.array([[4,2,1,1,1,2,3,2,1], [100,2,1,1,1,2,3,2,1]])
example_measures = example_measures.reshape(len(example_measures),-1)  # reshaping the array because we are not 
#passing the "Id" attribute

prediction = model.predict(example_measures)
print(prediction)
```
